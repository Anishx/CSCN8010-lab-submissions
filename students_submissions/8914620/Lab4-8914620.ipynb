### **Practical Lab 4 - Multivariate Linear and Polynomial Regression, and Evaluation using R-Squared, MAPE and MAE.**


```python
import pandas as pd
import numpy as np
import sklearn.datasets as datasets
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error

from sklearn.preprocessing import PolynomialFeatures
```

```python
X_diabetes, y_diabetes = datasets.load_diabetes(as_frame = True, scaled = False, return_X_y = True)
display(X_diabetes)
```

```python
display(y_diabetes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_diabetes, y_diabetes, test_size = 0.25, random_state = 4241)
model1 = LinearRegression()
model1.fit(X_train, y_train)

print("Linear Regression Intercept : ",model1.intercept_)
print("Coffecients : ",model1.coef_)

prediction = model1.predict(X_test)
```

### **Polynomial regression** 
#### Run a polynomial regression of the 2nd degree on the BMI feature alone

```python
polyTrain = PolynomialFeatures(degree = 2, include_bias = False)
LargeTrain2 = polyTrain.fit_transform(X_train)
polyTest = PolynomialFeatures(degree = 2, include_bias = False)
LargeTest2 = polyTest.fit_transform(X_test)
LargeTrain2 = pd.DataFrame(LargeTrain2, columns = polyTrain.get_feature_names_out())
display("LargeTrain2 :::",LargeTrain2)
LargeTest2 = pd.DataFrame(LargeTest2, columns = polyTest.get_feature_names_out())
display("LargeTest2",LargeTest2)

bmiXTrain2= LargeTrain2[['bmi','bmi^2']]
bmiXTest2= LargeTest2[['bmi','bmi^2']]
```

```python
bmiModel2 = LinearRegression()
bmiModel2.fit(bmiXTrain2,y_train)
print("\n BMI degree 2 Intercept ::::",bmiModel2.intercept_)
print("\n BMI degree 2 - Coefficient for BMI & BMI^2 ::::",bmiModel2.coef_)
```

#### **Run a polynomial regression of the 2nd degree on all features**
This case is with 2 features considered, BMI and BMI^2 and a intercept. 

```python
fullModel2 = LinearRegression()
fullModel2.fit(LargeTrain2, y_train)
print("\n All Features Degree 2 Intercept ::::",fullModel2.intercept_)
print("\n All Features Degree 2 Coefficient::::",fullModel2.coef_)
```

#### **Compare the three models by looking at R-squared, MAPE and MAE. Explain what the values mean for a non-expert and add your insight about the values of each model. Note: You can add any further comparisons and code (this is not necessary for a perfect score, but will be reviewed and evaluated)**

```python
print("All Features- Degree 1")  
AllTrainDeg1 = model1.predict(X_train)          
AllTestDeg1 = model1.predict(X_test)
# Train
print("R2TrainDegree1",r2_score(y_train, AllTrainDeg1))
print("MAETrainDegree1",mean_absolute_error(y_train, AllTrainDeg1))
print("MAPTrainDegree1", mean_absolute_percentage_error(y_train, AllTrainDeg1))
# Test
print("R2TestDegree1", r2_score(y_test, AllTestDeg1))
print("MAETestDegree1",  mean_absolute_error(y_test, AllTestDeg1))
print("MAPTestDegree1",mean_absolute_percentage_error(y_test, AllTestDeg1),"\n")

print("BMI feature alone- Degree 2")  
BMIpredictionTrainDegree2 = bmiModel2.predict(bmiXTrain2) 
BMIpredictionTestDegree2 = bmiModel2.predict(bmiXTest2)
# Train 
print("BMIR2TrainDegree2 ",r2_score(y_train, BMIpredictionTrainDegree2))
print("BMIMAETrainDegree2", mean_absolute_error(y_train, BMIpredictionTrainDegree2))
print("BMIMAPTrainDegree2", mean_absolute_percentage_error(y_train, BMIpredictionTrainDegree2))
# Test
print("BMIR2TestDegree2 = ",r2_score(y_test, BMIpredictionTestDegree2))
print("BMIMAETestDegree2 =", mean_absolute_error(y_test, BMIpredictionTestDegree2))
print("BMIMAPTestDegree2 =", mean_absolute_percentage_error(y_test, BMIpredictionTestDegree2),"\n")

print("All Features - Degree 2")  
AllTrainDeg2 = fullModel2.predict(LargeTrain2)  
AllTestDeg2 = fullModel2.predict(LargeTest2)
# Train 
print("R2TrainDegree2 = ",r2_score(y_train, AllTrainDeg2))
print("MAETrainDegree2 =", mean_absolute_error(y_train, AllTrainDeg2))
print("MAPTrainDegree2 =", mean_absolute_percentage_error(y_train, AllTrainDeg2))
# Test
print("R2TestDegree2 = ",r2_score(y_test, AllTestDeg2))
print("MAETestDegree2 =", mean_absolute_error(y_test, AllTestDeg2))
print("MAPTestDegree2 =", mean_absolute_percentage_error(y_test, AllTestDeg2),"\n")

```
